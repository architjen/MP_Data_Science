### Using the score_classifier function with dataset, classfier and labels

I'll bring some changes to the function **score_classifier**
- Compare multiple classification model and their recall performance
- Plot their recall across different splits, along with standard deviation  

**NOTE**: The functionality of original function is still respected, no changes in that
- KFold validation technique with 3 splits is used here
- The random state is set to 50
- The data is shuffled 
- The recalls are averaged over 3 splits to have the final performance 

```python
# different models
models = []
models.append(('LR', LogisticRegression( solver='lbfgs')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC( gamma='scale')))
```

```python
def score_classifier(dataset, models, labels):
    '''
    Params: 
        dataset: type(numpy ndarray), training data
        models: type(list), list of all the models to try
        labels: type(numpy ndarray), the labels
        
    Steps: What happens here basically
        - For each model it performs the 3 Kfold CV
        - appends the recall to the results
        - Prints avg recall and st deviation
        - plots all the recall for different models in the end
        
    Return:
        results: type(list), contains the recall performace of all the models
    '''
    results = [] # evaluate each model in turn
    names = []
    scoring = 'recall' #as we want to calculate recall
    for name, model in models:
        kfold = KFold(n_splits=3, random_state=50, shuffle=True)
        cv_results = cross_val_score(model, dataset, labels, cv=kfold, scoring=scoring)
        results.append(cv_results)
        names.append(name)
        print(f'Model: {name} | Avg Recall: {round(cv_results.mean(),3)} | Std: {round(cv_results.std(),3)}')

    # boxplot algorithm comparison
    fig = plt.figure(figsize=(15,10))
    fig.suptitle('Different algorithm Comparison based on Recall')
    ax = fig.add_subplot(111)
    plt.boxplot(results,showmeans=True, showfliers=False)
    ax.set_xticklabels(names)
    plt.show()
    
    return results
```

```python
r = score_classifier(dataset, models, labels)
```

What we can see from here is:
- LR gave the highest recall here
- We would still select SVM because: not only it performed better overall but also it gives the lowest variance amongst all the classifier 
- We can even select LR and SVM both here and proceed further, but for simplicity I'll only consider SVM 

So, whats the point of doing this?
- We benchmark all the different models 
- Select the best (or two) amongst all of them
- Fine-tune the model(s) further on to improve the performance and create a final model file

Fine tuning can be done by grid search or scikit optimize. We'll select GridSearch to fine tune our model  
By default, the GridSearchCV uses a 5-fold cross-validation, so we dont have to worry about validating using Kfold technique

```python
params = dict()
params['C'] = [0.1, 1, 10]
params['gamma'] =  [1, 0.1, 0.01]
params['degree'] = [1,2,3,4]
params['kernel'] = ['poly', 'rbf', 'sigmoid']
```

```python
#kfold = KFold(n_splits=3, random_state=50, shuffle=True)
search = GridSearchCV(SVC(), params, scoring='f1') 
search.fit(dataset, labels)
```

**NOTE**: What fit() does here is bit more complex, so first of all it goes over the loop of validation data to find the best parameter. Once it has the best parameters, it runs fit() on the entire data without CV and returns the single model which can be considered as the best model coming from the best parameters

```python
search
```

```python
# print how our model looks after hyper-parameter tuning
print(search.best_estimator_)
```

```python
predictions = search.predict(scaler.transform(x_test))
```

```python
print(recall_score(y_test, predictions))
```
